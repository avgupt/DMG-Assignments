```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data1 = pd.read_csv("cancerData.csv")
data1.head()
```

```python
data1.shape
```

```python
data1 = data1.replace('?',np.NaN)
```

```python
data1 = data1.dropna( axis = 1, thresh = 0.5*len(data1))
data1
```

```python
data1.nunique(axis=0)
```

```python
data1.isna().sum()
```

```python
data1.dtypes
```

```python
data1 = data1.apply(pd.to_numeric)
data1.dtypes
```

```python
import seaborn as sns

fig, ax = plt.subplots(figsize=(10,10))         # Sample figsize in inches

ax = sns.heatmap(data1.corr(), annot=False, ax = ax, linewidths=1)
```

```python
countLabel1 = data1[data1['Biopsy'] ==1]
countLabel0 = data1[data1['Biopsy'] ==0]
print( countLabel1['Biopsy'].count() )
print( countLabel0['Biopsy'].count() )
```

```python
no_of_columns = (data1.shape[1])
data1 = data1.dropna(axis=0,  thresh = 0.6 * no_of_columns)
data1
```
